```python
import pandas as pd
import re
import nltk
from gensim.models import Word2Vec
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
data = pd.read_csv("Reviews.csv", on_bad_lines='skip')
print(data.columns)

texts = data['Text'].dropna().tolist()
```

```python
def preprocess(i):
    i = i.lower()
    i = re.sub(r'[^a-zA-Z\s]', '', i)
    words = nltk.word_tokenize(i)
    return words
```

```python
corpus = [preprocess(t) for t in texts]
print("Sample:", corpus[0])
```

```python
wtov = Word2Vec(
    corpus,
    vector_size=10,
    window=5,
    min_count=2,
    workers=4,
    sg=1
)

wtov.save("word2vec.model")
print("Done!")
```
